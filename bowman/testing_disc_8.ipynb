```python
def mark_points(trin, points_set, rgbcolor=(0, 0, 0)):
    trin_copy = trin
    for tri_id, pt_coords in points_set:
        trin_copy = trin_copy.mark_point(tri_id, pt_coords, rgbcolor)
    return trin_copy

preserved_pts = points_preserved(prymreal, veech_elem, {(5, (1/2, 1/2, 0))})
mark_points(prymreal, preserved_pts).plot().show()
```
